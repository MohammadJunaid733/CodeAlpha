```python
def generate_range_with_step(start, end, step=2):
    """
    Generate a list of numbers within a specified range with a given step size.

    Args:
        start (int): The starting number of the range (inclusive).
        end (int): The ending number of the range (inclusive).
        step (int): The step size between numbers (default is 2).

    Returns:
        List[int]: A list of numbers within the specified range with the given step size.
    """
    if start > end:
        return []
    
    num_list = []
    current = start
    while current <= end:
        num_list.append(current)
        current += step
    
    return num_list

# Example usage:
start = 0
end = 10
step = 2
result = generate_range_with_step(start, end, step)
print(result)
```
